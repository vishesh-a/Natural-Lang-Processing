```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://www.bhaskar.com/national/news/opposition-leaders-meet-ec-demands-50-vvpats-tallied-with-evms-01485180.html?ref=ht"
```

```python
web_page = urlopen(url)
```

```python
soup = BeautifulSoup(web_page, 'html.parser')
```

```python
for paragraph in soup.findAll():
        print (paragraph.text)
```

```python
print('\n')
```
